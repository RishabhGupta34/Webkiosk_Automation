```python
#URL FOR WEBKIOSK
url="https://webkiosk.thapar.edu/index.jsp?_ga=2.256712781.730507136.1514315530-602781231.1510251130"

from selenium import webdriver
browser=webdriver.Chrome()

browser.get(url)

#For the username
username=browser.find_element_by_name("MemberCode")
username.send_keys('101603278') #Enter the username/roll no.

#For the password
input_pass=browser.find_element_by_name("Password")
from getpass import getpass
password=getpass()
input_pass.send_keys(password) #A prompt occurs to enter the password

from selenium.webdriver.common.keys import Keys
input_pass.send_keys(Keys.ENTER)

frames=browser.find_elements_by_tag_name('frame')

link=frames[1].get_attribute('src')

browser.get(link)

divs=browser.find_elements_by_tag_name('div')

divs[5].click()

aa=browser.find_elements_by_tag_name('a')

for i in aa:
    if i.get_attribute("title")=="View Event Subject Marks":
        att=i

url=att.get_attribute('href')

browser.get(url)

option=browser.find_elements_by_tag_name('option')

for i in option:
    if i.get_attribute("value")=="1718ODDSEM": #For saving the marks of 1718ODDSEM
        i.click()
        break

#att.click()

inputs=browser.find_elements_by_tag_name('input')

for i in inputs:
    if i.get_attribute("value")=="Show":
        i.click()
        break

#atts.click()

html=browser.page_source

from bs4 import BeautifulSoup

soup=BeautifulSoup(html,"html5lib")

tablelist=soup.findAll('table')

table=tablelist[2]

tr=table.findAll('tr')

rows=[]

for t in tr:
    tds=t.findAll('td')
    row=[]
    for td in tds:
        row.append(td.text)
    rows.append(row)

import csv
with open("records.csv","w") as csvfile: 
    #For writing the marks in records.csv file in the directory where this python file executes
    csvwrite=csv.writer(csvfile)
    csvwrite.writerow(rows[0])
    csvwrite.writerows(rows[1:])
    
browser.close()
```
